```python
#Aluno: Victor Hugo dos Reis Bazzo
#MBA em Business Intelligence & Analytics e Big Data

#Importação de Bibliotecas
from datetime import datetime 

#Lendo arquivo TXT.
ref_arquivo = open("C:/Users/victo/Documents/Projetos/Aula01Python/ConversaWhats.txt","r", encoding = 'utf-8')
linha = ref_arquivo.readline()

#Variáveis de agrupamento e auxiliares
datas = []
diassemana = []
horarios = []
pessoas = []
msgs = []
erros = []
qtdhrsemaior = 0
qtdhrsemenor = 100
qtdemsgsmaior = 0
qtdediamaior = 0

#Inserindo cada tipo de dado em seus vetores/variaveis
while linha:
    valores = linha.split(" ", 6)
    try:
        datas = datas + valores[0:1]    
        #datas = datas + datetime.strptime(valores[0:1],'%d/%m/%y')
        horarios = horarios + valores[1:2]
        pessoas = pessoas + valores[5:6]
        msgs = msgs + valores[6:]
    except ValueError:
        erros.append('Erros de Formatacao de Data')
    linha = ref_arquivo.readline()

#Verificando os horários com mais ou menos mensagens
for valor in horarios:
    #print(horarios.count(valor))
    if (horarios.count(valor) > qtdhrsemaior):
        qtdhrsemaior = horarios.count(valor)
        valorhrsmaior = valor
    elif (horarios.count(valor) < qtdhrsemenor):
        qtdhrsemenor = horarios.count(valor)
        valorhrsmenor = valor
print("O horário das : ",valorhrsmaior," possui o maior indices de mensagens. Com a quantidade de: " ,qtdhrsemaior)
print("O horário das : ",valorhrsmenor," possui o menor indices de mensagens. Com a quantidade de: " ,qtdhrsemenor)


#Verificando qual palavra é mais utilizada
for valormsgs in msgs:
    #print(horarios.count(valor))
    if (msgs.count(valormsgs) > qtdemsgsmaior):
        qtdemsgsmaior = msgs.count(valormsgs)
        valormsgmaior = valormsgs
#Imprimindo a palavra com maior incidencia.
print("A palavra ",valormsgmaior," possui o maior indices de mensagens. Com a quantidade de: " ,qtdemsgsmaior)

#Tratando as datas para pegar o dia da semana e criar um vetor com os dias
for valordatas in datas:
    try:
        #teste = datetime.strptime(valordatas,'%d/%m/%y')
        teste = datetime.strptime(valordatas,'%d/%m/%Y')
        teste1 = teste.strftime('%A')
        #print(teste1)
        diassemana.append(teste1) 
        #if (datas.count(valordatas) > qtdediamaior):
        #    qtdediamaior = teste
        #print(teste.strftime('%A'))
    except ValueError:
        erros.append('Erros de Formatacao de Data')
    linha = ref_arquivo.readline()

#Calculando a maior incidencia de dia da semana.
for dias in diassemana:
    if (diassemana.count(dias) > qtdediamaior):
        qtdediamaior = diassemana.count(dias)
        valordiamaior = dias
print("O dia da Semana com maior incidência é: ",valordiamaior,". Com a quantidade de: " ,qtdediamaior)
        
ref_arquivo.close()
```
